### Exercice 1 :

Exportation de chaque types de variables (nominale, ordonnée et quantitative), visualisation en accord avec la taxonomie de Mackinlay et une visualisation ne respectant pas cette taxonomie.

### Importation des librairies necessaires ainsi que des données que nous allons étudier.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image, ImageDraw

from plotnine import ggplot, aes, geom_bar, geom_line, geom_path

nominal = pd.read_csv('data/Nominal_values.csv', sep = ';')
ordinal = pd.read_csv('data/Ord_Values.csv', sep = ';')
quantitative = pd.read_csv('data/Qaunti_Values.csv', sep = ';')
```

On charge les captures d'écran des graphiques pertinent pour chaques variables ainsi que pour les grphiques non-pertinent en y ajoutant du texte.

Nous avons ici :
 - le sexe des enfants à Angers de 2013 à 2019 pour la variable nominal
 - le nombre code INSEE des communes (regroupé en "IRIS") en fonction de l'exploitation de leur appellation "IRIS" (1 : exploitable, 2 : partiellement exploitable, 3: très peu exploitable , 4: non-exploitable) pour la variable ordonnée.
 - l'évolution du nombre de naissance par années pour la variable quantitative.
 
 Note : 
 Pour plus de détails sur l'IRIS : https://www.insee.fr/fr/metadonnees/definition/c1523 (ref. Site de l'INSEE)
 Les images sont disponible dans le dossier.

```python
ord_img = Image.open('images/Capture_ordonn_ok.PNG')
draw_ord = ImageDraw.Draw(ord_img)
draw_ord.text((500, 50),"Représentation correct de la variable ordonnée",(0,0,0))

quant_img = Image.open('images/Capture_quanti.PNG')
draw_quant = ImageDraw.Draw(quant_img)
draw_quant.text((500, 50),"Représentation correct de la variable quantitative",(0,0,0))

nom_img = Image.open('images/Capture_nominal_ok.PNG')
draw_nom = ImageDraw.Draw(nom_img)
draw_nom.text((200, 150),"Représentation correct de la variable nominale",(0,0,0))

nom_img.show()
ord_img.show()
quant_img.show()
```

```python
bad_ord_img = Image.open('images/Capture_ordonné_false.PNG')
draw_bad_ord = ImageDraw.Draw(bad_ord_img)
draw_bad_ord.text((500, 50),"Représentation incorrect d'une variable ordonnée",(0,0,0))

bad_quant_img = Image.open('images/Capture_quanti_false.PNG')
draw_bad_quant = ImageDraw.Draw(bad_quant_img)
draw_bad_quant.text((500, 5),"Représentation incorrect d'une variable quantitative",(0,0,0))

bad_nom_img = Image.open('images/Capture_nom_false.PNG')
draw_bad_nom = ImageDraw.Draw(bad_nom_img)
draw_bad_nom.text((583, 25),"Représentation incorrect d'une variable nominale",(0,0,0))

bad_nom_img.show()
bad_ord_img.show()
bad_quant_img.show()
```
